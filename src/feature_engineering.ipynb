```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/cleaned/cleaned_food_orders.csv')
df.head(5)
```

```python
customer_summary = df.groupby('Customer_ID').agg({
    'final_order_value': ['sum', 'mean'],
    'Order_ID': 'count'
})
```

```python
customer_summary.columns = ['total_spent', 'avg_order_value', 'total_orders']
customer_summary = customer_summary.reset_index()
```

```python
customer_summary['spending_category'] = pd.qcut(
    customer_summary['total_spent'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
customer_summary['frequency_category'] = pd.qcut(
    customer_summary['total_orders'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```
