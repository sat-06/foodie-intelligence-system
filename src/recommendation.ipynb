```python
def recommend_food(df, hour, city=None, top_n=5):
    filtered = df[df['order_hour'] == hour]

    if city is not None:
        city_data = filtered[filtered['City'] == city]
        if len(city_data) > 0:
            filtered = city_data

    if len(filtered) == 0:
        filtered = df

    restaurants = filtered['Resto_name'].value_counts().head(top_n)
    items = filtered['Items_in_order'].value_counts().head(top_n)

    return restaurants, items
```

```python
restaurants, items = recommend_food(df, hour=20, city='Pune')

print("Recommended Restaurants:")
print(restaurants)

print("\nRecommended Items:")
print(items)
```
