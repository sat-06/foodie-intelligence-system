```python
df.rename(columns={'Restaurant ID':'Resto_ID','Restaurant name':'Resto_name','Order ID':'Order_ID','Order Placed At':'Order_Placed_At','Order Status':'Order_status','Items in order':'Items_in_order','Discount construct':'Discount_construct','Bill subtotal':'Bill_subtotal','Packaging charges':'Packaging_charges','Restaurant discount (Promo)':'Restaurant_discount_(Promo)','Restaurant discount (Flat offs, Freebies & others)':'Restaurant_discount_(Flat offs, Freebies & others)','Gold discount':'Gold_discount','Brand pack discount':'Brand_pack_discount','Cancellation / Rejection reason':'Cancellation/Rejection_reason','Restaurant compensation (Cancellation)':'Restaurant_compensation_(Cancellation)','Restaurant penalty (Rejection)':'Restaurant_penalty_(Rejection)','KPT duration (minutes)':'KPT_duration_(minutes)','Rider wait time (minutes)':'Rider_wait_time_(minutes)','Order Ready Marked':'Order_Ready_Marked','Customer complaint tag':'Customer_complaint_tag','Customer ID':'Customer_ID'}, inplace=True)
df.columns
```

```python
df['Instructions']=df['Instructions'].fillna('No Instrutions')
df['Dicount_construct']=df['Discount_construct'].fillna('No Discount')
df['Rating']=df['Rating'].fillna('Not Rated')
df.drop('Cancellation/Rejection_reason',axis=1,inplace=True)
df.drop('Restaurant_compensation_(Cancellation)',axis=1,inplace=True)
df.drop('Restaurant_penalty_(Rejection)',axis=1,inplace=True)
df['KPT_duration_(minutes)']=df['KPT_duration_(minutes)'].fillna(df['KPT_duration_(minutes)'].median())
df['Rider_wait_time_(minutes)']=df['Rider_wait_time_(minutes)'].fillna(df['Rider_wait_time_(minutes)'].median())
df.drop('Customer_complaint_tag',axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Discount_construct'] = df['Discount_construct'].replace(['', 'null', 'None'], None)
```

```python
df['Discount_construct'] = df['Discount_construct'].fillna('No Discount')
```

```python
df.isnull().sum()
```

```python
df['Distance'] = df['Distance'].str.replace('km', '')
df['Distance'] = df['Distance'].str.strip()
```

```python
df['Distance'] = pd.to_numeric(df['Distance'], errors='coerce')
```

```python
df['Order_Placed_At']=pd.to_datetime(df['Order_Placed_At'])
numeric_cols = [
    'Distance',
    'Bill_subtotal',
    'Packaging_charges',
    'Restaurant_discount_(Promo)',
    'Restaurant_discount_(Flat offs, Freebies & others)',
    'Gold_discount',
    'Brand_pack_discount',
    'Total',
    'KPT_duration_(minutes)',
    'Rider_wait_time_(minutes)'
]
for col in numeric_cols:
    if col != 'Distance':
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.dtypes
```

```python
df['Order_Placed_At'] = pd.to_datetime(df['Order_Placed_At'], errors='coerce')
df['Order_Placed_At'].dtype
```

```python
df['order_date'] = df['Order_Placed_At'].dt.date
df['order_hour'] = df['Order_Placed_At'].dt.hour
df['order_day_name'] = df['Order_Placed_At'].dt.day_name()
df['order_month'] = df['Order_Placed_At'].dt.month
df['is_weekend'] = df['Order_Placed_At'].dt.dayofweek >= 5
```

```python
df['total_discount']=(
    df['Restaurant_discount_(Promo)']+df['Restaurant_discount_(Flat offs, Freebies & others)']+df['Gold_discount']+df['Brand_pack_discount']
)
```
